```python
import json
```

```python
data = [json.loads(line) for line in open('data_wrong.jsonl')]
```

```python
user2items = {}
for item in data:
    uid, iid = item['USER_ID'], item['ITEM_ID']
    if uid not in user2items: user2items[uid] = []
    user2items[uid].append(iid)
```

```python
data = []
for uid in user2items:
    data.append({'user': uid, 'ids': user2items[uid]})
```

```python
with open('data_correct.jsonl', 'w') as fout:
    for item in data:
        fout.write(json.dumps(item) + '\n')
```
